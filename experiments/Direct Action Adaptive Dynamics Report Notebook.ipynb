```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def payoff(R,S,T,P,y,x):
    return R*x*y + S*(1-x)*y + T*x*(1-y) + P*(1-x)*(1-y)

def invasion_fitness(R,S,T,P,y,x):
    return payoff(R,S,T,P,y,x) - payoff(R,S,T,P,x,x)

def invasion_fitness_derivative(R,S,T,P,x):
    return (R-S-T+P)*x + (S-P)

x = []
dx = []
dx2 = []
axis_x =np.arange(-0.5, 1.5, 0.1).tolist()
axis = np.zeros(len(axis_x))
for i in range(11):
    x.append(round(i*0.1,1))
    dx.append(invasion_fitness_derivative(4,1,3,2,x[-1]))
    dx2.append(invasion_fitness_derivative(6,1,3,2,x[-1]))
plt.plot(x, dx, c = 'g', label = "matrix = [4,1;3,2]")
plt.plot(x, dx2, c = 'b', label = "matrix = [6,1;3,2]")
plt.plot(axis_x, axis, c = "black")
plt.title("Derivative of the Invasion Fitness Function of a Rare Mutant Strategy")
plt.xlabel("Dominant belief")
plt.ylabel("Derivative")
plt.xlim(0,1)
plt.legend()
plt.show()
```

```python
def payoff_alpha(R,S,T,P,y,x,alpha):
    return alpha*payoff(R,S,T,P,y,y) + (1-alpha)*payoff(R,S,T,P,y,x)

def invasion_fitness_alpha(R,S,T,P,y,x,alpha):
    return payoff_alpha(R,S,T,P,y,x,alpha) - payoff_alpha(R,S,T,P,x,x,alpha)

def invasion_fitness_derivative_alpha(R,S,T,P,x,alpha):
    return (1+alpha)*(R-S-T+P)*x + (S+alpha*T-P-alpha*P)

x = []
dx0 = []
dx05 = []
dx1 = []
axis_x =np.arange(-0.5, 1.5, 0.1).tolist()
axis = np.zeros(len(axis_x))
for i in range(11):
    x.append(round(i*0.1,1))
    dx0.append(invasion_fitness_derivative_alpha(4,1,3,2,x[-1],0))
    dx05.append(invasion_fitness_derivative_alpha(4,1,3,2,x[-1],0.5))
    dx1.append(invasion_fitness_derivative_alpha(4,1,3,2,x[-1],1))
plt.plot(x, dx0, c = 'g', label = "alpha = 0.0")
plt.plot(x, dx05, c = 'b', label = "alpha = 0.5")
plt.plot(x, dx1, c = 'r', label = "alpha = 1.0")
plt.plot(axis_x, axis, c = 'black')
plt.title("Derivative of the Invasion Fitness Function of a Rare Mutant Strategy")
plt.xlabel("Dominant belief")
plt.ylabel("Derivative")
plt.xlim(0,1)
plt.legend()
plt.show()
```
