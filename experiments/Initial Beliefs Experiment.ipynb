```python
import model
import numpy as np;
import matplotlib.pyplot as plt;
```

```python
def experiment(number_of_agents, Rf, Sf, Tf, Pf, Ro, So, To, Po, tag0_in, tag0_out, tag1_in, tag1_out, number_of_steps, rounds_per_step, selection_intensity, perturbation_probability, perturbation_scale, data_recording, game_name, performance_independent):
    a_model = model.Model(number_of_agents, Rf, Sf, Tf, Pf, Ro, So, To, Po, tag0_in, tag0_out, tag1_in, tag1_out)
    a_model.run_simulation(number_of_steps, rounds_per_step, selection_intensity, perturbation_probability, perturbation_scale, data_recording, performance_independent)
    
    y1 = np.array(a_model.avg_ingroup_0_time_series)
    y2 = np.array(a_model.avg_ingroup_1_time_series)
    y3 = np.array(a_model.avg_outgroup_0_time_series)
    y4 = np.array(a_model.avg_outgroup_1_time_series)
    plt.plot(y1, label="ingroup 0")
    plt.plot(y2, label="ingroup 1")
    plt.plot(y3, label="outgroup 0")
    plt.plot(y4, label="outgroup 1")
    plt.legend()
    plt.title("{}: Average Belief at Time Step".format(game_name) )
    plt.show()
```

```python
# Parameters

number_of_agents = 10;
number_of_steps = 1000;
rounds_per_step = 100;
selection_intensity = 0.1;
perturbation_scale = 0.05;
performance_independent = True;
perturbation_probability = 0.1

# PD =  [-1, -3, 0, -2]
# SH = [4, 1, 3, 2]
```

```python
# Initial Beliefs 
tag0_in = 1
tag0_out = 0
tag1_in = 1
tag1_out = 0

# PD game
for _ in range(3):
    experiment(number_of_agents, -1, -3, 0, -2, -1, -3, 0, -2, tag0_in, tag0_out, tag1_in, tag1_out, number_of_steps, rounds_per_step, selection_intensity, perturbation_probability, perturbation_scale, False, "PD", performance_independent)

# SH game
for _ in range(3):
    experiment(number_of_agents, 4, 1, 3, 2, 4, 1, 3, 2, tag0_in, tag0_out, tag1_in, tag1_out, number_of_steps, rounds_per_step, selection_intensity, perturbation_probability, perturbation_scale, False, "SH", performance_independent)
```
