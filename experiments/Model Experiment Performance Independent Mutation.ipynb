# Experiment: Performance Independent Mutation

Uses the version of the model with simple learning where in the new generation each agent has a chance to have their beliefs perturbed.

Tests model on Prisoner's Dilemma and Stag Hunt with varying perturbation probability.

Reports graphs of average payoffs of each group.

```python
from src.python import model;
import numpy as np;
import matplotlib.pyplot as plt;
```

```python
def experiment(number_of_agents, Rf, Sf, Tf, Pf, Ro, So, To, Po, number_of_steps, rounds_per_step, selection_intensity, perturbation_probability, perturbation_scale, data_recording, game_name, performance_independent):
    a_model = model.Model(number_of_agents, Rf, Sf, Tf, Pf, Ro, So, To, Po)
    a_model.run_simulation(number_of_steps, rounds_per_step, selection_intensity, perturbation_probability, perturbation_scale, data_recording, performance_independent)
    
    y1 = np.array(a_model.avg_payoff_0_time_series)
    y2 = np.array(a_model.avg_payoff_1_time_series)
    plt.plot(y1, label="0")
    plt.plot(y2, label="1")
    plt.legend()
    plt.title("{}: Average Payoff at Time Step with {} Perturbation Probability".format(game_name, perturbation_probability) )
    plt.show()
```

```python
# Parameters

number_of_agents = 10;
number_of_steps = 1000;
rounds_per_step = 100;
selection_intensity = 0.1;
perturbation_scale = 0.05;
performance_independent = True;

# PD =  [-1, -3, 0, -2]
# SH = [4, 1, 3, 2]
```

```python
# Test with 0.1 perturbation probability
perturbation_probability = 0.1

# PD game
experiment(number_of_agents, -1, -3, 0, -2, -1, -3, 0, -2, number_of_steps, rounds_per_step, selection_intensity, perturbation_probability, perturbation_scale, False, "PD")

# SH game
for _ in range(5):
    experiment(number_of_agents, 4, 1, 3, 2, 4, 1, 3, 2, number_of_steps, rounds_per_step, selection_intensity, perturbation_probability, perturbation_scale, False, "SH")
```

```python
# Test with 0.5 perturbation probability
perturbation_probability = 0.5

# PD game
experiment(number_of_agents, -1, -3, 0, -2, -1, -3, 0, -2, number_of_steps, rounds_per_step, selection_intensity, perturbation_probability, perturbation_scale, False, "PD")

# SH game
for _ in range(5):
    experiment(number_of_agents, 4, 1, 3, 2, 4, 1, 3, 2, number_of_steps, rounds_per_step, selection_intensity, perturbation_probability, perturbation_scale, False, "SH")
```
