```python
# Basic Once-Off Stag Hunt

def payoff(R,S,T,P,y,x):
    return R*x*y + S*(1-x)*y + T*x*(1-y) + P*(1-x)*(1-y)

def invasion_fitness(R,S,T,P,y,x):
    return payoff(R,S,T,P,y,x) - payoff(R,S,T,P,x,x)

def invasion_fitness_derivative(R,S,T,P,x):
    return (R-S-T+P)*x + (S-P)

x = []
dx = []
dx2 = []
axis_x =np.arange(-0.5, 1.5, 0.1).tolist()
axis = np.zeros(len(axis_x))
for i in range(11):
    x.append(round(i*0.1,1))
    dx.append(invasion_fitness_derivative(4,1,3,2,x[-1]))
    dx2.append(invasion_fitness_derivative(6,1,3,2,x[-1]))
plt.plot(x, dx, c = 'g', label = "matrix = [4,1;3,2]")
plt.plot(x, dx2, c = 'b', label = "matrix = [6,1;3,2]")
plt.plot(axis_x, axis, c = "black")
plt.title("Derivative of the Invasion Fitness Function of a Rare Mutant Strategy")
plt.xlabel("Dominant belief")
plt.ylabel("Derivative")
plt.xlim(0,1)
plt.legend()
plt.show()


y = []
ifit = []
ifit2 = []
ifit3 = []
ifit4 = []
ifit5 = []
axis_x =np.arange(-0.5, 1.5, 0.1).tolist()
axis = np.zeros(len(axis_x))
for i in range(11):
    y.append(round(i*0.1,1))
    ifit.append(invasion_fitness(4,1,3,2,y[-1],0.1))
    ifit3.append(invasion_fitness(4,1,3,2,y[-1],0.9))
    ifit4.append(invasion_fitness(4,1,3,2,y[-1],0.3))
    ifit5.append(invasion_fitness(4,1,3,2,y[-1],0.7))
plt.plot(y, ifit, c = 'g', label = "x = 0.1")
plt.plot(y, ifit4, c = 'orange', label = "x = 0.3")
plt.plot(y, ifit5, c = 'purple', label = "x = 0.7")
plt.plot(y, ifit3, c = 'r', label = "x = 0.9")
plt.plot(axis_x, axis, c = "black")
plt.title("Invasion Fitness of a Rare Mutant Strategy y\n against Resident Strategy x")
plt.xlabel("Mutant belief")
plt.ylabel("Invasion Fitness")
plt.xlim(0,1)
plt.legend()
plt.show()

print(ifit2)
```

```python
# Once-Off Stag Hunt with Ingroup and Outgroup strategies 

def payoff_inout(R,S,T,P,y,x,beta):
    return beta*payoff(R,S,T,P,y[0],x[0]) + (1-beta)*payoff(R,S,T,P,y[1],x[1])

def invasion_fitness_inout(R,S,T,P,y,x,beta):
    return payoff_inout(R,S,T,P,y,x,beta) - payoff_inout(R,S,T,P,x,x,beta)

def invasion_fitness_derivative_inout(R,S,T,P,x,beta):
    return (beta*((R-S-T+P)*x[0] +(S-P)), (1-beta)*((R-S-T+P)*x[1] + (S-P)))

print(invasion_fitness_inout(4,1,3,2,[0.6,0.6],[0.2, 0.2],0.0))
```

```python
# Once-Off Stag Hunt with Ingroup and Outgroup strategies and Alpha 

def payoff_full(R,S,T,P,y,x,beta,alpha):
    return alpha*payoff(R,S,T,P,y[0],y[0]) + (1-alpha)*(beta*payoff(R,S,T,P,y[0],x[0])+(1-beta)*payoff(R,S,T,P,y[1],x[1]))

def invasion_fitness_full(R,S,T,P,y,x,beta,alpha):
    return payoff_full(R,S,T,P,y,x,beta,alpha) - payoff_full(R,S,T,P,x,x,beta,alpha)


print(invasion_fitness_full(4,1,3,2,[0.6,0.6],[0.2, 0.2],0.5,0.7))
```
